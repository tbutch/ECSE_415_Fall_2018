# Laplacian of Gaussian
Question 2.2.1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image as greyscale
imgWoman = cv2.imread("woman.png", cv2.IMREAD_GRAYSCALE)

# Apply laplacian filter to image of woman
imgLaplacian = cv2.Laplacian(imgWoman, cv2.CV_32F, ksize=5)

# Compute zero crossings in the laplacian image
# Create zeros array
height, width = imgWoman.shape
zeros = np.zeros((height, width))

# Iterate through pixels in the laplacian image, find values
# Where the laplacian is zero to find "Zero Crossings" in the image
for x in range(height):
    for y in range(width):
        # If laplacian is zero at particular x,y
        if(imgLaplacian[x][y] == 0):
            # Set zeros array to 255 to signify an edge
            zeros[x][y] = 255
        # Cover negative next to positive in x --> We imagine that there is a zero crossing
        # between the two pixels. Also covers edge of image cases
        elif(imgLaplacian[x][y] < 0 and (imgLaplacian[min(x+1, height-1)][y] > 0 or imgLaplacian[max(x-1,0)][y] > 0)):
            zeros[x][y] = 255
        # Same situation, but in y
        elif(imgLaplacian[x][y] < 0 and (imgLaplacian[x][min(y+1, width -1)] > 0 or imgLaplacian[x][max(y-1,0)] > 0)):
            zeros[x][y] = 255


img = [imgWoman, imgLaplacian, zeros]
titles = ["Original Image", "Laplacian of Gaussian", "Edges"]
# Display images
plt.figure(num=None, figsize=(50,50), dpi=100)
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(img[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
plt.figure(num=None, figsize=(10,10), dpi=50)
plt.subplot(1,1,1), plt.imshow(zeros, 'gray'), plt.title("Edges / Zero Crossings")
plt.show()
```
