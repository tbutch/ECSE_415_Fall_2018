# Sharpening
Question 1.2.1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage.measure

# Load image
imgRice = cv2.imread("rice.png")

# No need to convert image to greyscale, already b&w

# Apply box filter to image 
box_filter = np.ones((5,5), np.float32) / (5*5)
box_filtered_img = cv2.filter2D(imgRice, -1, box_filter)

# Compute Details with box filter and create sharpened image
details_box = imgRice - box_filtered_img
sharpened_box = imgRice + details_box

# Apply gaussian filter to image with varX=2, varY=2
gaussian_blurred = cv2.GaussianBlur(imgRice, (5,5), 2^2)
details_gaussian = imgRice - gaussian_blurred
sharpened_gaussian = imgRice + details_gaussian

# Show images
plot = plt.figure(num=None, figsize=(10,10), dpi=100)
# Add original image
ax = plot.add_subplot(1,3,1), plt.imshow(imgRice), plt.title("Original")
# Add box filtered sharpened image
ax2 = plot.add_subplot(1,3,2), plt.imshow(sharpened_box), plt.title("Box Sharpening")
# Add gaussian filtered image
ax3 = plot.add_subplot(1,3,3), plt.imshow(sharpened_gaussian), plt.title("Gaussian Sharpening")

plt.show()

```

Question 1.2.2

When comparing both sharpening methods, "better" is a very subjective term when looking at an image. However, it is expected that the gaussian filter provide better blurring to the image, in the sense that hard edges are better preserved due to its weighted average nature. 

Such, it is expected that having better preserved the sharp edges, the Gaussian sharpening is the be better than the box sharpening. Here are my observations when viewing both images:
The box sharpened image presents a large contrast between the rice grains and the gray background. However, when observing the original image, it appears that there has been an intensity shift, such that the grey portions of the rice are now closer to white. Comparatively in the gaussian filter, the contrast is not as huge, but there is a color fidelity that is not present in the box sharpening. Also, due to the higher preservation of sharp edges with the gaussian filter, the edges are just as sharp than the box filter. Such, the gaussian filter provides a better sharpening effect.
