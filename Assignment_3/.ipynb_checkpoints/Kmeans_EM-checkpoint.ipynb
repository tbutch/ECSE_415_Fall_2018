```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image
img = cv2.imread("home.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

#imgList = img.reshape((img.shape[0] * img.shape[1], 3))

# Define number of clusters
k=3

# KMeans function
def KMeans_segmentation(inImg, clusterNum):
    
    # Copy input image
    outImg = inImg.copy()  
    # reshape the image to be a list of pixels with depth 3
    inList = inImg.reshape((inImg.shape[0]* inImg.shape[1]),3)

    super_pix_list = []
    for y in range(inImg.shape[0]):
        for x in range(inImg.shape[1]):
            super_pix = [y,x,inImg[y][x][0],inImg[y][x][1],inImg[y][x][2]]
            super_pix_list.append(super_pix)
    
    # Define Centroids array
    centroids = [None] * clusterNum
    
    # Pick random Points in list to act as cluster centers and store them in centroids
    for i in range(clusterNum):
        posit = np.random.randint(1, len(super_pix_list)+1)
        centroids[i] = super_pix_list[posit]
    
    # Measure euclidian distance for each of the pixels in the image and assign them to a centroid
    # Repeat this loop 10 times
    for num in range(10):
        # Create 2D list of points associated to each centroid
        # Define list of pixels
        listOfPixels = []
        for center in centroids:
            centerList = []
            listOfPixels.append(centerList)
            
        for pixelVal in super_pix_list:
            distance = []
            for i in range(clusterNum):
                distance.append(0)
            pixelRGB = np.array([pixelVal[2],pixelVal[3],pixelVal[4]])
            
            # Compute distance of pixel to each centroid
            for i in range(clusterNum):
                centroidsRGB = np.array([centroids[i][2], centroids[i][3], centroids[i][4]])
                rSSD = float(pixelRGB[0]) - centroidsRGB[0]
                gSSD = float(pixelRGB[1]) - centroidsRGB[1]
                bSSD = float(pixelRGB[2]) - centroidsRGB[2]
                measured_dist = (rSSD*rSSD + gSSD*gSSD + bSSD*bSSD)
                distance[i] = measured_dist
            # Find minimum distance in distance list
            min_dist = distance[0]
            closest_center = 0
            for i in range(clusterNum):
                if(distance[i] < min_dist):
                    min_dist = distance[i]
                    closest_center = i
            # Append pixel to closest center list
            listOfPixels[closest_center].append(pixelVal)

        # Select new centers for each list of points associated to a center
        # For every list, first compute 1st SSD
        RGB_Mean = []
        for i in range(clusterNum):
            RGB_Mean.append([0,0,0])
        
        # Compute Mean R,G and B values for each cluster
        for i in range(clusterNum):
            # For each cluster, compute mean of R,G,B values and store in RGB_Mean
            running_sum_r = 0
            running_sum_g = 0
            running_sum_b = 0
            for w in range(len(listOfPixels[i])):
                running_sum_r += listOfPixels[i][w][2]
                running_sum_g += listOfPixels[i][w][3]
                running_sum_b += listOfPixels[i][w][4]
            
            RGB_Mean[i][0] = (running_sum_r/len(listOfPixels[i]))
            RGB_Mean[i][1] = (running_sum_g/len(listOfPixels[i]))
            RGB_Mean[i][2] = (running_sum_b/len(listOfPixels[i]))
        
        # Then, compute SSD b/w other pixels and mean. Minimize that distance
        # in order to find closest pixel to the mean of the cluster
        SSD = []
        computed_ssd = []
        for i in range(clusterNum):
            SSD.append(0)
            computed_ssd.append(0)
        
        # Compute 1st SSD from first pixel for each cluster to set benchmark
        for i in range(clusterNum):
            potential_cent = listOfPixels[i][0]
            rSSD = (potential_cent[2]) - RGB_Mean[i][0]
            gSSD = (potential_cent[3]) - RGB_Mean[i][1]
            bSSD = (potential_cent[4]) - RGB_Mean[i][2]
            SSD[i] = (rSSD*rSSD + gSSD*gSSD + bSSD*bSSD)
            centroids[i] = potential_cent
        
        for i in range(clusterNum):
            
            for w in range(len(listOfPixels[i])):
                potential_centroid = listOfPixels[i][w]
                # Compute SSD from potential new center
                rSSD = (potential_centroid[2]) - RGB_Mean[i][0]
                gSSD = (potential_centroid[3]) - RGB_Mean[i][1]
                bSSD = (potential_centroid[4]) - RGB_Mean[i][2]
                point_ssd = (rSSD*rSSD + gSSD*gSSD + bSSD*bSSD)
                
                if(point_ssd < SSD[i]):
                    SSD[i] = point_ssd
                    centroids[i] = potential_centroid
            
    # After finding final center points and list associated to each center point, make final image
    # Assign pixels associated to the center the value of the center
    # Copy initial image to get correct dimensions
    outImage = inImg.copy()
    for i in range(clusterNum):
        for j in range(len(listOfPixels[i])):
            pixel = listOfPixels[i][j]
            row_pos = pixel[0]
            col_pos = pixel[1]
            resulting_pixel = np.array([centroids[i][2],centroids[i][3],centroids[i][4]])
            outImage[row_pos][col_pos] = resulting_pixel
    
    outImage = np.uint8(outImage)
    return outImage
    
    
# segmented_image = img.copy()
# segmented_image2 = np.zeros(img.shape)
#print(img[0][0])

segmented_image = KMeans_segmentation(img,2)
segmented_image2 = KMeans_segmentation(img,3)
segmented_image3 = KMeans_segmentation(img,20)

```

### Implement the EM algorithm using only the numpy library. You can use opencv and matplotlib libraries only to read and display images. Apply it to the given image with 2 and 3 Gaussian components. Display the resulting segmented images for 10 iterations in both cases. (20 points)

NOTE: My algorithm is based on the methodology for EM explained Here: https://www.python-course.eu/expectation_maximization_and_gaussian_mixture_models.php?fbclid=IwAR0kRYvMC9EIRggV8DAKM-MREa9gJBZn2ubjKE-NsMD84g9ZwxccAZzVIWk

```python


    
def probability_map(clusterNum, dataList, mean, covariance, weight):
    """
    Function:    probability_map(dataList, mean, covariance, weight)
    Description: Function to intialize random gaussian clusters and parameters
    Parameters:  numClusters - Number of Gaussians
                 dataList ---- List of data in the image, containing [x,y,r,g,b] info
                 mean -------- Mean matrix of the gaussian clusters
                 covariance -- Covariance matrix of the gaussian clusters
                 weight ------ Weight matrix of the gaussian clusters
    
    Returns:     p_map ------- probability map for each cluster
    """
    prob_map = np.zeros((len(dataList),clusterNum))
    for i in range(len(dataList)):
        
        for j in range(clusterNum):
            # Calculate the probabilitu that point xi is in cluster K
            # store this value in prob_map[i][j]
            
            cov_curr = covariance[j] #??????
            
            cov_inv = np.linalg.inv(cov_curr)
            # Remove X,Y parts used only to make image
            current_data = np.asarray([dataList[i][2], dataList[i][3], dataList[i][4]])
            #current_mean = np.asarray([mean[j][2],mean[j][3],mean[j][4]])
            current_mean = mean[j]
            
            diff = current_data - current_mean
            # constant portion of the Multivariate gaussian
            const = 1/(np.sqrt(np.linalg.det(cov_curr))* np.power(2*np.pi, 3/2))
            # Exponential portion of Multivariate gaussian
            expon = np.exp((-1/2)*np.dot((np.transpose(diff)), np.dot(cov_inv, diff)))
            # Multivariate gaussian for datapoint xi
            gauss = const*expon
            prob_map[i][j] = gauss
            
        # Numerator for probability
        denominator = sum(prob_map[i])
        prob_map[i] = [prob_map[i][c]/denominator for c in range(clusterNum)]
        
        
    return prob_map
    
    
def gaussianInit(clusterNum,dataList):
    """
    Function:    gaussianInit(numClusters, dataList)
    Description: Function to intialize random gaussian clusters and parameters
    Parameters:  numClusters - Number of Gaussians
                 dataList ---- List of data in the image, containing [x,y,r,g,b] info
    
    Returns:     mean -------- Mean matrix of the gaussian clusters
                 covariance -- Covariance matrix of the gaussian clusters
                 weight ------ Weight matrix of the gaussian clusters
    """
    # numclusters x 5 mean ie gaussian center point
    mean = np.zeros((clusterNum,3))
    # numclusters x 3 x 3
    covariance = np.zeros((clusterNum,3,3))
    # weight = 1 x numclusters
    weight = np.zeros((clusterNum))
    
    # Randomly select datapoints within list to be mean points
    for i in range(clusterNum):
        randpt = dataList[np.random.randint(0,len(dataList) + 1)]
        mean[i] = np.asarray([randpt[2],randpt[3],randpt[4]])
    covariance[:] = np.eye(3)
    weight[:] = 1/clusterNum

    
    return mean,covariance,weight
    
    
    
def EM_Segmentation(inImg, clusterNum, iterNum):
    """
    Function:    EM_Segmentation(inImg,clusterNum,iterNum)
    Description: Function to calculate Expectation Maximisation using gaussian mixture models
    Parameters:  inImg ------ Input image that needs to be segmented using EM, NEEDS to be normalized (divided by 255)
                 clusterNum - Number of desired gaussian matrices
                 iterNum ---- Number of iterations to re-compute the gaussian means
                 
    Returns:     outImage --- Segmented image to be displayed
    
    """
    # Copy input image
    outImg = inImg.copy()  
    # reshape the image to be a list of pixels with depth 3
    inList = inImg.reshape((inImg.shape[0]* inImg.shape[1]),3)

    super_pix_list = []
    for y in range(inImg.shape[0]):
        for x in range(inImg.shape[1]):
            super_pix = [y,x,inImg[y][x][0],inImg[y][x][1],inImg[y][x][2]]
            super_pix_list.append(super_pix)
    
    # Define initial gaussian parameters
    mean,covariance,weight = gaussianInit(clusterNum,super_pix_list)
    
    # Repeat this loop iterNum times
    for num in range(iterNum):
        
        ##### E stage #####
        # Calculate probability map
        print("Computing probability map,  ", num)
        prob_map = probability_map(clusterNum, super_pix_list, mean, covariance, weight)
        print("Probability map computed, ", num)
        
        ##### M stage #####
        # initialize new empty mean and cov arr
        mean_new = np.zeros((clusterNum,3))
        cov_new = np.zeros((clusterNum,3,3))
        
        print("Calculating sum of probabilites,", num)
        # Calculate total sum of probabilities to compute new mean
        prob_total = np.zeros((clusterNum))
        for i in range(clusterNum):
            for j in range(len(super_pix_list)):
                prob_total[i] += prob_map[j][i]
        
        # Compute new weight matrix
        print("Computing new weight matrix, ", num)
        weight = prob_total/len(super_pix_list)
        
        
        # Compute new mean
        print("Computing new mean points, ", num)
        for i in range(clusterNum):
            for j in range(len(super_pix_list)):
                rgb_data  = np.asarray([super_pix_list[j][2],super_pix_list[j][3], super_pix_list[j][4]])
                mean_new[i] += prob_map[j][i] * rgb_data
            mean_new[i] /= prob_total[i]
            
        # New covariance matrix compute
        print("Computing new covariance matrix, ", num)
        for i in range(clusterNum):
            for j in range(len(super_pix_list)):
                rgb_data = np.asarray([super_pix_list[j][2],super_pix_list[j][3],super_pix_list[j][4]])
                diff = (rgb_data - mean_new[i])[np.newaxis]
                cov_new[i] += prob_map[j][i]*(np.dot(np.transpose(diff),diff))
            cov_new[i] /= prob_total[i]
               
        mean = mean_new
        covariance = cov_new
    # After finding final center points and list associated to each center point, make final image
    # Assign pixels associated to the center the value of the center
    # Create 2D list of points associated to each centroid
    # Define list of pixels
    listOfPixels = []
    for i in range(clusterNum):
        centerList = []
        listOfPixels.append(centerList)
        
    for i in range(len(super_pix_list)):
        # Take each point, find max probability for each point and assign to a list
        max_idx = np.argmax(prob_map[i])
        listOfPixels[max_idx].append(super_pix_list[i])
    
    # Copy initial image to get correct dimensions
    print("Preparing final image:")
    outImage = inImg.copy()
    for i in range(clusterNum):
        for j in range(len(listOfPixels[i])):
            pixel = listOfPixels[i][j]
            row_pos = pixel[0]
            col_pos = pixel[1]
            resulting_pixel = np.array([mean[i][0]*255,mean[i][1]*255,mean[i][2]*255])
            outImage[row_pos][col_pos] = resulting_pixel
    
    outImage = np.uint8(outImage)
    return outImage

img255 = img/255
seg_img = EM_Segmentation(img255,3,10)
seg_img2 = EM_Segmentation(img255,2,10)

```

### 1.1.3 Under what data distribution conditions would K-means and EM give the same solution, provided similar initializations? (5 points)

K-Means and EM would give the same solution under certain circumstances. First of all, the initialisation must be the same: The randomly selected points at the start of the algorithm greatly influence the ending of the algorithm. That being said, the main difference between the K-Means algorithm and the EM algorithm is the assumption made about the shape of the pixel clusters.
First of all, it should be noted that K-Means is a simplification of EM, such that the shape of the pixel clusters are assumed to be circular/spherical. On the other hand, the EM algorithm does not conform to this assumption, meaning that the shape of the cluster can fall within an ellipsoidal shape described by a multivariate gaussian function. Thus, the only way that K-Means and EM could come to the same solution is if the data clusters in EM turn out to be exactly circular.
