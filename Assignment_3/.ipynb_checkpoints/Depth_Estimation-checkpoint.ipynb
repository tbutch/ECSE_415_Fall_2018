2.1 Depth Estimation from Stereo Images (13 points)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import skimage
```

```python
import sys
import re
from struct import *
import numpy as np
from matplotlib import pyplot as plt

def read_pfm(file):
    # Adopted from https://stackoverflow.com/questions/37073108/how-to-read-pfm-files-provided-with-middlebury-dataset?rq=1
    with open(file, "rb") as f:
        # Line 1: PF=>RGB (3 channels), Pf=>Greyscale (1 channel)
        type = f.readline().decode('latin-1')
        if "PF" in type:
            channels = 3
        elif "Pf" in type:
            channels = 1
        else:
            sys.exit(1)
        # Line 2: width height
        line = f.readline().decode('latin-1')
        width, height = re.findall('\d+', line)
        width = int(width)
        height = int(height)

        # Line 3: +ve number means big endian, negative means little endian
        line = f.readline().decode('latin-1')
        BigEndian = True
        if "-" in line:
            BigEndian = False
        # Slurp all binary data
        samples = width * height * channels;
        buffer = f.read(samples * 4)
        # Unpack floats with appropriate endianness
        if BigEndian:
            fmt = ">"
        else:
            fmt = "<"
        fmt = fmt + str(samples) + "f"
        img = unpack(fmt, buffer)
        img = np.flipud(np.array(img).reshape(height, width))
    return img
```

2.1.1 Estimate depth using the stereo image pair. Show the results of varying the following parameters: blockSize in range [5; 10; 15], numDisparities in range [16; 32; 48]. You will have a total of 9 depth-maps, each corresponding to one combination of blocksize and numDisparities. (5 points)

```python
img_0 = cv2.imread("im0.png")
img_1 = cv2.imread("im1.png")

img_0_smol = cv2.resize(img_0, (0,0), fx=0.3, fy=0.3) 
img_1_smol = cv2.resize(img_1, (0,0), fx=0.3, fy=0.3) 

img_0_gray = cv2.cvtColor(img_0_smol, cv2.COLOR_BGR2GRAY)
img_1_gray = cv2.cvtColor(img_1_smol, cv2.COLOR_BGR2GRAY)

block_size = [5,9,15]
num_disparities = [16,32,48]

disparity_maps = []
titles = []

for i in range(len(block_size)):
    for j in range(len(num_disparities)):
        
        # Create stereoBM object with the correct # of disparities and block size
        stereo = cv2.StereoBM_create(numDisparities=num_disparities[j], blockSize=block_size[i])
        disparity = stereo.compute(img_0_gray,img_1_gray)
        
        # Append the calculated disparity map to the list of images
        disparity_maps.append(disparity)
        
        # Make image titles
        title = "Disparity Map: numDisparities = %i, blockSize = %i" % (num_disparities[j],block_size[i])
        titles.append(title)
        
# Display disparity maps
plt.figure(figsize = (20,20))
for i in range(len(disparity_maps)):
    disp = disparity_maps[i]
    plt.subplot(3,3,i+1),plt.imshow(disp,'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
    
    
# depth_img0[depth_img0==np.inf] = 0
# plt.imshow(depth_img0); plt.axis('off'); plt.show()
```

2.1.2 For each depth-map, compute the pixel-wise squared difference error with respect to the ground-truth. (2 points)

```python
depth_img0 = read_pfm('disp0.pfm')

depth_img0_smol = cv2.resize(depth_img0, (0,0), fx=0.3, fy=0.3)
# replace all Inf depth with 0 <==> undefined depth
depth_img0_smol[depth_img0_smol==np.inf] = 0
#plt.imshow(depth_img0_smol); plt.axis('off'); plt.show()

square_difference = []
img_ssd = []
for i in range(9):
    img_ssd.append(abs(disparity_maps[i] - depth_img0_smol))

plt.figure(figsize = (20,20))
for i in range(len(img_ssd)):
    disp = img_ssd[i]
    plt.subplot(3,3,i+1),plt.imshow(disp,'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()

for disparity in disparity_maps:
    diff = 0
    for y in range(disparity.shape[0]):
        for x in range(disparity.shape[1]): 
            disp_pix = disparity[y][x]/255
            ground_truth_pix = depth_img0_smol[y][x]/255
            if(disp_pix < 0):
                disp_pix = 0
            elif(disp_pix > 255):
                disp_pix = 255
            elif(np.isnan(disp_pix)):
                disp_pix = 255
                
            if(ground_truth_pix < 0):
                ground_truth_pix = 0
            elif(ground_truth_pix > 255):
                ground_truth_pix = 255
            elif(np.isnan(ground_truth_pix)):
                ground_truth_pix = 255   
            
            # Normalise pixels
            ground_truth_pix = ground_truth_pix/255
            disp_pix = disp_pix/2
            calc_diff = float(disp_pix) - ground_truth_pix
            diff = diff + (calc_diff*calc_diff)

    square_difference.append(diff)
            
for i in range(len(square_difference)):
    print("Computed SSD for image", i+1, ":", square_difference[i])

```

### 2.2.3 Comment on the error-maps for each combination of parameters. Specifically, mention the effects of varying the parameters on the error in the region of high/low texture, edge, large/small depths. (6 points)

The error maps were computed by subtracting the ground truth image from the disparity map, in order to visually see if the image is close to the ground truth. Also, I computed the overall SSD for each image in order to get an idea as to if the whole image tends to match the ground truth.

With that being said, the effect of the parameters will largely depend on the image being used. Here, the image that I use has a lot of foreground and background details, which makes small block sizes show a lot of noise and errors, as can be observed in the first row of images with block size 5. The noise is especially exacerbated in high texture regions, where, for example, the body of the motorcycle which has a very textured, glossy look to it.

Increasing the block size tended to yield better results in the disparity map, and such a lower error in the error maps above. In fact, the image with the best result is the disparity map with a block size of 15 and the number of disparities at 16 (empirically, it has the lowest SSD at 31984.xx). This is because the larger block size helps to bring out the general shape of the images instead of the noisier parts of the image.

Varying the number of disparities helped with the depth perception of the image. This is an entirely qualitative analysis, but as the number of disparities increased (especially in the bottom row) the perception of depth/ what is foreground/background is exaggerated, especially if we look at the bottom-center and bottom-right images, in which the optimal block size is used.
